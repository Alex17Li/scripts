```python
import math
kernel_size = 10
sigma = 8

# Create a x, y coordinate grid of shape (kernel_size, kernel_size, 2)
x_cord = torch.arange(kernel_size)
x_grid = x_cord.repeat(kernel_size).view(kernel_size, kernel_size)
y_grid = x_grid.t()
xy_grid = torch.stack([x_grid, y_grid], dim=-1)

mean = (kernel_size - 1)/2.
variance = sigma**2.

gaussian_kernel = (1./(2.*math.pi*variance)) *\
                  torch.exp(
                      -torch.sum((xy_grid - mean)**2., dim=-1) /\
                      (2*variance)
                  )
# Make sure sum of values in gaussian kernel equals 1.
gaussian_kernel = gaussian_kernel / torch.sum(gaussian_kernel)

# Reshape to 2d depthwise convolutional weight
gaussian_kernel = gaussian_kernel.view(1, kernel_size, kernel_size)
gaussian_kernel = torch.nn.Parameter(gaussian_kernel.repeat(1, 1, 1, 1))
gaussian_kernel.requires_grad = False
```

# We need to do the forward transform first instead of applying the reverse transform like the literature so that the saliency values correspond to the correct place

```python
class NonUniformSamplingLayer(torch.nn.Module):
    def __init__(
            self,
        ):
        super().__init__()
        # assert downsample_rows % n_row_divisions == 0
        # assert downsample_cols % n_col_divisions == 0
        # self.row_expand_copies = downsample_rows // n_row_divisions
        # self.col_expand_copies = downsample_cols // n_col_divisions

    def compute_transforms_1d_(self, saliency: torch.Tensor) -> Tuple[torch.Tensor, torch.Tensor]:
        B, L = saliency.shape
        device = saliency.device
        # We construct 2 mappings. (For both row and col)
        # Map from the highres domain to the downsampled range (1 pixel per downsampled pixel)
        # Map from the downsampled domain to the highres range (1 pixel per highres pixel)
        # We know the intended range (1 point per pixel center), and must compute the domains
        highres_range = torch.arange(-1, 1, 2 /  L, device=device)  + 1 / L
        downsampled_range = torch.arange(-1, 1, 2 / L, device=device) + 1 / L

        # Get the midpoint of each section
        cum_saliency = torch.cumsum(saliency, 1)
        downsampled_domain_unnormalized = cum_saliency - saliency / 2
        # Normalize to [-1, 1] by dividing by the sum and scaling
        downsampled_domain = 2 * downsampled_domain_unnormalized / cum_saliency[:, -1:] - 1
        highres_range_pad = torch.cat([-torch.ones(1, device=device), highres_range, torch.ones(1, device=device)])
        # downsampled_range_r_pad = torch.cat([-torch.ones(1), downsampled_range_r, torch.ones(1)])
        downsampled_domain_pad = torch.cat([-torch.ones(B, 1, device=device), downsampled_domain, torch.ones(B, 1, device=device)], dim=1)
        inverse_transform_index = torch.sum(downsampled_domain_pad[:, :, None] < downsampled_range[None, None, :], dim=1)
        downsampled_lo = torch.gather(downsampled_domain_pad, 1, inverse_transform_index - 1)
        downsampled_hi = torch.gather(downsampled_domain_pad, 1, inverse_transform_index)
        highres_domain_hi = highres_range_pad[inverse_transform_index]
        highres_domain_lo = highres_range_pad[inverse_transform_index - 1]
        weight = (downsampled_range - downsampled_lo) / (downsampled_hi - downsampled_lo)
        highres_domain = torch.lerp(highres_domain_lo, highres_domain_hi, weight)
        return downsampled_domain, highres_domain

    def get_transforms(self, saliency: torch.Tensor) -> Tuple[torch.Tensor, torch.Tensor]:
        """
        saliency: floats with value >0 corresponding to the desired zoom for a tile of the image.
        
        returns:
            nonuniform_downsampling_grid: Apply to a batch of images to get a nonuniformly downsampled image of shape [self.downsample_rows, self.downsample_cols]
            inverse_transform_grid: : Invert the distortions created by the transform to get a tensor of shape [B, _, H, W]
        """
        saliency = saliency[:, 0, :, :]
        B, H, W = saliency.shape

        # It is the importance of the rows at a given height. For a single step in the highres range, we expect
        # that it will be zoomed by a factor corresponding to it's saliency. Thus, when mapping backwards from the highres
        # range to the downsampled domain, we can find the points in the mapping using the saliency.
        saliency_of_cols = torch.max(saliency, dim=1).values
        saliency_of_rows = torch.max(saliency, dim=2).values
        downsampled_domain_r, highres_domain_r = self.compute_transforms_1d_(saliency_of_rows)
        downsampled_domain_c, highres_domain_c = self.compute_transforms_1d_(saliency_of_cols)

        inverse_grid = []
        for i in range(B):
            inverse_grid.append(torch.stack(torch.meshgrid(downsampled_domain_c[i], downsampled_domain_r[i], indexing='xy'), dim=0))
        inverse_grid = torch.stack(inverse_grid, dim=0)
        
        nonuniform_downsampling_grid = []
        for i in range(B):
            nonuniform_downsampling_grid.append(torch.stack(torch.meshgrid(highres_domain_c[i], highres_domain_r[i], indexing='xy'), dim=0))
        nonuniform_downsampling_grid = torch.stack(nonuniform_downsampling_grid, dim=0)
    
        return nonuniform_downsampling_grid, inverse_grid, downsampled_domain_c, downsampled_domain_r

    def apply_transform(self, sampling_grid: torch.Tensor, image: torch.Tensor, scale_factor: float) -> torch.Tensor:
        target_width = round(image.shape[2] * scale_factor)
        target_height = round(image.shape[3] * scale_factor)
        if target_width != sampling_grid.shape[2] or target_height != sampling_grid.shape[3]:
            sampling_grid = torch.nn.functional.interpolate(sampling_grid, (target_width, target_height), align_corners=False, mode='bilinear')
        sampling_grid_reshape = sampling_grid.permute(0,2,3,1)
        # align_corners=False is required to be correct/consistent with the logic of using the centers of pixels
        return F.grid_sample(image, sampling_grid_reshape, align_corners=False)
```

# OK let's try a simple, fixed transform. If we fix the transform globally then we should be able to invert it quickly

```python
class NonUniformSamplingLayer(torch.nn.Module):
    def __init__(
            self,
            downsample_rows: int = 32,
            downsample_cols: int = 32,
            n_row_divisions: int = 16,
            n_col_divisions: int = 16,
        ):
        super().__init__()
        assert downsample_rows % n_row_divisions == 0
        assert downsample_cols % n_col_divisions == 0
        self.row_expand_copies = downsample_rows // n_row_divisions
        self.col_expand_copies = downsample_cols // n_col_divisions

    def performn_nonuniform_s(self, image_fullres: torch.Tensor, saliency: torch.Tensor) -> torch.Tensor:
        """
        image shape: [B, C, H, W]
        saliency shape: floats with value >0 corresponding to the desired zoom for a tile of the image.
        Shape [B, 1, n_row_divisions, n_col_divisions]
        
        returns:
            nonuniformlysampled_image shape [B, C, downsample_rows, downsample_cols]
        """
        # For simpler computation of the reverse transform, we do the rows and columns independently.
        # We determine the total size of a row proportional to the most salient elemnt in it.
        x_weights = torch.repeat_interleave(torch.max(saliency, dim=2), repeats=self.row_expand_copies, dim=2)
        y_weights = torch.repeat_interleave(torch.max(saliency, dim=3), repeats=self.col_expand_copies, dim=2)

        # Get the midpoint of each section
        xvals_unnormalized = torch.cumsum(x_weights, 1) - x_weights / 2
        yvals_unnormalized = torch.cumsum(y_weights, 1) - y_weights / 2

        # Normalize to [-1, 1]
        xvals = 2 * xvals_unnormalized / torch.sum(x_weights, 1) - 1
        yvals = 2 * yvals_unnormalized / torch.sum(y_weights, 1) - 1
        
        # Perform the remapping
        grids = []
        for i in range(xvals.shape[0]):
            grids.append(torch.meshgrid(xvals[i], yvals[i], indexing='xy').permute([1, 2, 0]))
        nonuniformlysampled_image = grid_sample(image_fullres, grids, align_corners=False)
        return nonuniformlysampled_image
```
