```python
def load_object(row):
    folder_path  = directory + "/processed/images/" + row['id'] + "/"
    file_name = [f for f in os.listdir(folder_path) if 'stereo' in f][0]
    return np.load(folder_path + file_name)

def show_images(df):
    assert len(df) < 100, "Bro, too many images"
    n, m = len(df), 3
    fig, ax = plt.subplots(n, m, figsize=(m * 3.2, n * 4), squeeze=False)
    for i, (_, row) in enumerate(df.iterrows()):
        obj = load_object(row)
        im_l = normalize_image(obj['left'], row['hdr_mode'])
        ax[i][0].set_title(row['id'])
        ax[i][0].imshow(im_l)

        im_r = normalize_image(obj['right'], row['hdr_mode'])
        ax[i][1].set_title('right')
        ax[i][1].imshow(im_r)

        im_depth = obj['point_cloud'][..., 2] / 100
        ax[i][2].set_title('depth')
        ax[i][2].imshow(im_depth)
```

```python
show_images(stereo_df)
```

```python
import pyvista
pyvista.set_jupyter_backend('trame')
pyvista.start_xvfb()
```

```python
grid = pyvista.ImageData(dimensions=(3, 3, 1))
ugrid = grid.cast_to_unstructured_grid()
```

```python
pl = pyvista.Plotter()
pl.add_mesh(ugrid, show_edges=True, line_width=5)
label_coords = ugrid.points + [0, 0, 0.02]
point_labels = [f'Point {i}' for i in range(ugrid.n_points)]
pl.add_point_labels(label_coords, point_labels,
                    font_size=25, point_size=20)
cell_labels = [f'Cell {i}' for i in range(ugrid.n_cells)]
pl.add_point_labels(ugrid.cell_centers(), cell_labels, font_size=25)
pl.camera_position = 'xy'
pl.show()
```
